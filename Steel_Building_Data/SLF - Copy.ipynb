```python
import storeyloss
import pandas as pd
import json

# Load the necessary data from CSV files
component_data = pd.read_csv("MyInventory_NSD/My_Inventory_CASE1.csv")


# Initialize and configure the SLF model
model = storeyloss.SLF(
    component_data,
    'psd',    
    do_grouping=True,
)

# Generate loss functions
out = model.generate_slfs()

# Export output to JSON
model.export_to_json(out, "MyInventory_NSD\\output_1.json")


# # Load JSON data from the file
# with open('MyInventory_NSD/output_1.json', 'r') as file:
#     data = json.load(file)

# # Extract data for '2.edp_range' and '2.slf'
# edp_range = data['2']['edp_range']
# slf = data['2']['slf']

# # Create a DataFrame
# df = pd.DataFrame({
#     'EDP Range': edp_range,
#     'SLF': slf
# })

# # Save the DataFrame to a CSV file
# df.to_csv('MyInventory_NSD/output_1CCCC.csv', index=False)

# Assuming 'out' can be directly converted to DataFrame, if not already one
if not isinstance(out, pd.DataFrame):
    out = pd.DataFrame(out)

# Export the same output to CSV
out.to_csv("D:\\PhD\\StoryLossFunctions\\MyInventory_NSD\\output_1.csv", index=False)


```

```python
import json
import pandas as pd

# Load JSON data from the uploaded file
with open('D:\\PhD\\StoryLossFunctions\\MyInventory_NSD\\output_3.json', 'r') as file:
    data = json.load(file)

# Attempt to convert to DataFrame
try:
    df = pd.json_normalize(data)
except Exception as e:
    print("Error in converting to DataFrame:", e)
    df = None

# If DataFrame creation is successful, display the first few rows and info
if df is not None:
    print("DataFrame Info:")
    print(df.info())
    print("\nFirst few rows of the DataFrame:")
    print(df.head())
else:
    print("Data needs further processing or visualization")

import matplotlib.pyplot as plt

# Extracting data for the first component
edp_range_3 = df.at[0, '2.edp_range']
slf_values_3 = df.at[0, '2.slf']
```
